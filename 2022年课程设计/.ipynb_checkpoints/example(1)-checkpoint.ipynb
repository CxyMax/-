# 数据重整

## 离散变量的重编码

由于数据集中有很多离散型变量，这样的字符变量是不能直接用于建模的，需要对这些变量进行重编码，关于重编码的方法有多种，如将字符型的值转换为整数型的值、哑变量处理（0-1变
量）、One-Hot独热编码（类似于哑变量）等。在本案例中，将采用“字符转数值”的方法对离散型变量进行重编码，具体可以通过下面的代码实现：


```python
for feature in bank.columns:
    if bank[feature].dtype == 'object':
        bank[feature] = pd.Categorical(bank[feature]).codes
bank.head(10)
```

结果就是对字符型离散变量的重编码效果，所有的字符
型变量都变成了整数型变量，如job、marital、housing、loan等，

## 删除冗余变量或明显无关的变量
数据集中，id：每一条记录一个id，与最终订购没有关系。可以将id变量从数据集中删除。

另外，pdays的四分位数（0，999，999，999，999）：previous的四分位数（0，0，0，7），说明变量相同值太多，信息量很少，也可以删掉。

```python
bank.drop(['id'], axis = 1, inplace = True)
bank.head()
```

呈现的就是经处理“干净”的数据集，所要预测的变量就是
income，该变量是二元变量，对其预测的实质就是对年收入水平的分类
（一个新样本进来，通过分类模型，可以将该样本分为哪一种收入水
平）。


## 数据拆分：训练集（0.7）+测试集（0.3）

基于上面的“干净”数据集，需要将其拆分为两个部分，一部分用于分类器模型的构建，另一部分用于分类器模型的评估，这样做的目的是避免分类器模型过拟合或欠拟合。如果模型在训练集上表现很好，而在测试集中表现很差，则说明分类器模型属于过拟合状态；如果模型在训练过程中都不能很好地拟合数据，那说明模型属于欠拟合状态。通常情况下，会把训练集和测试集的比例分配为75%和25%：


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(bank.loc[:,'age':'nr.employed'], 
                                                    bank['label'], train_size = 0.70, 
                                                    random_state = 54321)
print('训练数据集共有%d条观测' %X_train.shape[0])
print('测试数据集共有%d条观测' %X_test.shape[0])
```

```python
X_train
```

```python
y_train
```

# 数据建模

## k近邻模型

### 导入k近邻模型的类

对K近邻模型，这里直接调用sklearn子模块neighbors中的KNeighborsClassifier类，并且使用模型的默认参数，即让K近邻模型自动挑选最佳的搜寻近邻算法（algorithm='auto'）、使用欧氏距离公式计算样本间的距离（p=2）、指定未知分类样本的近邻个数为5（n_neighbors=5）而且所有近邻样本的权重都相等（weights='uniform'）。

注意：使用不同的样本间的距离计算公式、近邻个数分析结果异同。

```python
from sklearn.neighbors import KNeighborsClassifier
```

### 构建k近邻模型

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
print(knn)
```

### 训练集评估
#### 准确率

```python
print('KNN模型在训练集上的准确率：%f' %knn.score(X_train,y_train))
```

```python
knn_pred_X_train = knn.predict(X_train)
print(pd.crosstab(knn_pred_X_train, y_train, margins = True))
```

#### f1_score

```python
from sklearn import metrics
f1_knn_train = metrics.f1_score(knn_pred_X_train, y_train)
print('KNN模型在训练集上的f1_score：%f' %f1_knn_train)
```

### 测试集评估
#### 准确率

```python
print('KNN模型在测试集上的准确率：%f' %knn.score(X_test,y_test))
```

```python
knn_pred_X_test = knn.predict(X_test)
print(pd.crosstab(knn_pred_X_test, y_test, margins = True))
```

#### f1_score

```python
f1_knn_test = metrics.f1_score(knn_pred_X_test, y_test)
print('KNN模型在测试集上的f1_score：%f' %f1_knn_test)
```

### 过拟合判断

```python
print('KNN模型在训练集上的准确率:%f' %knn.score(X_train,y_train))
print('KNN模型在测试集上的准确率:%f' %knn.score(X_test,y_test))
```

```python
print('KNN模型在训练集上的f1_score: %f' % metrics.f1_score(knn_pred_X_train, y_train))
print('KNN模型在测试集上的f1_score: %f' % metrics.f1_score(knn_pred_X_test, y_test))
```

经过计算，KNN模型在训练集中的准确率为93.1%，但在测试集上的准确率90.4%，说明默认参数下的KNN模型基本不存在过拟合的风险。


## GBDT模型

对GBDT模型，可以调用sklearn子模块ensemble中的GradientBoostingClassifier类，同样先尝试使用该模型的默认参数，即让模型的学习率（迭代步长）为0.1（learning_rate=0.1）、损失函数使用的是对数损失函数（loss='deviance'）、生成100棵基础决策树（n_estimators=100），并且每棵基础决策树的最大深度为3（max_depth=3），中间节点（非叶节点）的最小样本量为2（min_samples_split=2），叶节点的最小样本量为1（min_samples_leaf=1），每一棵树的训练都不会基于上一棵树的结果（warm_start=False）。

### 导入GBDT模型的类

```python
from sklearn.ensemble import GradientBoostingClassifier
```

### 构建GBDT模型

```python
gbdt = GradientBoostingClassifier()
gbdt.fit(X_train, y_train)
print(gbdt)
```

### 训练集评估

#### 准确率

```python
print('GBDT模型在训练集上的准确率：%f' %gbdt.score(X_train,y_train))
```

```python
gbdt_pred_X_train = gbdt.predict(X_train)
print(pd.crosstab(gbdt_pred_X_train, y_train, margins = True))
```

#### f1-score

```python
f1_gbdt_train = metrics.f1_score(gbdt_pred_X_train, y_train)
print('GBDT模型在测试集上的f1_score：%f' %f1_gbdt_train)
```

### 测试集评估

#### 准确率

```python
print('GBDT模型在测试集上的准确率：%f' %gbdt.score(X_test,y_test))
```

```python
gbdt_pred_X_test = gbdt.predict(X_test)
print(pd.crosstab(gbdt_pred_X_test, y_test, margins = True))
```

#### f1_score

```python
f1_gbdt_test = metrics.f1_score(gbdt_pred_X_test, y_test)
print('GBDT模型在测试集上的f1_score：%f' %f1_gbdt_test)
```

### 过拟合判断

```python
print('GBDT模型在训练集上的准确率%f' %gbdt.score(X_train,y_train))
print('GBDT模型在测试集上的准确率%f' %gbdt.score(X_test,y_test))
```

```python
print('GBDT模型在训练集上的f1_score: %f' % metrics.f1_score(gbdt_pred_X_train, y_train))
print('GBDT模型在测试集上的f1_score: %f' % metrics.f1_score(gbdt_pred_X_test, y_test))
```

GBDT模型在训练集中的准确率为92.4%，但在测试集上的准确率91.5%，说明集成算法GBDT在测试集上的表现明显要比KNN算法优秀，这就是基于多棵决策树进行投票的优点。可以预见GBDT模型在预测集上应该有更好的结果，但差距不大。
